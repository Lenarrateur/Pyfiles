```python
%matplotlib inline

x = np.linspace(0,3*np.pi,500)

plt.plot(x,np.sin(x**2))
plt.title('A simple chirp')
```

```python
x = np.linspace (-2,2,101)
y=x**2
y2=x**4
plt.xlabel("x")
plt.ylabel("f(x)")
plt.plot(x,y, 'g', label = "x ^ 2")
plt.plot(x,y2,'ro', label = "x ^ 4")
plt.title('A simple chirp')
plt.legend()

#plt.savefig ("fig1.png")
plt.show()
```

```python

u = int(input ("Numbers of point ?"))
x = np.linspace (-2,2,u)
y = np.cos(2*np.pi*x)
plt.title('Body function 2pix')
plt.xlabel("x")
plt.ylabel("f(x)")
plt.plot(x,y)
plt.savefig ("cos2pix.png")
plt.show()
```

```python
u = int(input ("Numbers of point ?"))
x = np.linspace (-2,2,u)
y = np.cos(2*np.pi*x)
y2 =np.sin(2*np.pi*x)
plt.title('Body function 2pix')
plt.xlabel("x")
plt.ylabel("f(x)")
plt.plot(x,y,label = "cos")
plt.plot(x,y2,label = "sin")
plt.savefig ("trigonometric.png")
plt.legend()
plt.show()
```

```python

u = int(input ("Numbers of point ?"))
x = np.linspace (0,4,u)
y = (np.sin(np.pi*x))*(np.sin(np.pi*20*x)*np.exp(-x))
plt.xlabel("x")
plt.ylabel("f(x)")
plt.plot(x,y)
plt.savefig ("cos2pix.png")
plt.show()

```

```python
u = int(input ("Numbers of point ?"))
x = np.linspace (2,10,u)
nb = int(input("Number of Isotherm ?"))
for i in range (nb):
         v = int(input ("Temperature ?"))
         y = 0.08206*v/x
         plt.plot(x,y)
plt.xlabel("x")
plt.ylabel("f(x)")
plt.title('Isotherm (ideal gas)')
plt.savefig ("isotherm.png")
plt.show()
```

```python


u = int(input ("Numbers of point ?"))
v = int(input ("Initial Value:"))
w = int(input ("End value:"))
x0 = int(input ("xO:"))
S = float(input ("S:"))
x = np.linspace (v,w,u)
y = (1/((2*np.pi)**0.5)*np.exp((x-x0)**2/S**2/(-2)))
plt.xlabel("x")
plt.ylabel("f(x)")
plt.plot(x,y)
plt.savefig ("cos2pix.png")
plt.show()
```

```python
u = int(input ("Numbers of point ?"))
x = np.linspace (0,3,u)
y = np.exp(-x)
y2 =np.sin(np.pi*3*x)*np.exp(-x)
plt.title('Functions')
plt.xlabel("x")
plt.ylabel("f(x)")
plt.plot(x,y,label = "e-x")
plt.plot(x,y2,label = "sin(3pix)*e-x")
plt.savefig ("trigonometric.png")
plt.legend()
plt.show()
```

```python

u = int(input ("Numbers of point ?"))
t= int(input( "Number of functions"))
v = int(input ("Initial Value:"))
w = int(input ("End value:"))
x = np.linspace (v,w,u)
for i in range (t) :
    x0 = int(input ("xO:"))
    S = float(input ("S:"))
    y = (1/((2*np.pi)**0.5)*np.exp((x-x0)**2/S**2/(-2)))
    plt.plot(x,y)
plt.xlabel("x")
plt.ylabel("f(x)")
plt.plot(x,y)
plt.savefig ("gaussians.png")
plt.show()
```
